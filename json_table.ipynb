```python
import pandas as pd
```

```python
df = pd.read_csv("./data/links.csv")
df.shape
```

```python
df.tail()
```

```python
df1 = df[~df['Peacebuilding Interventions'].isnull()]
df1.tail()
```

```python
df1 = df1.groupby(['Tags','Peacebuilding Interventions','Type'])['Links'].apply(list).reset_index(name='Links')
df1
```

```python
import json

# Define a custom function to create a nested structure
def custom_nested_structure(row):
    return {row['Peacebuilding Interventions']: 
            {row['Tags']: 
             {'Links': row['Links'], 
              'Type': row['Type']}
              }}

# Apply the custom function to each row of the DataFrame
json_data_custom = df1.apply(custom_nested_structure, axis=1).tolist()

# Convert list of dictionaries to JSON
json_output = json.dumps(json_data_custom, indent=4)

print(json_output)

```

```python
with open("./data/links.json", "w") as file:
    file.write(json_output)
```
